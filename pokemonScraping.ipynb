```python
import requests, re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import lxml
```

```python
import warnings
warnings.filterwarnings('ignore')


def generateCharacterList(url):
    websiteURL =  requests.get(url)
    soup = BeautifulSoup(websiteURL.text)
    characterGallery = soup.findAll('table',{'align':'center'})
    characters = []

    for table in characterGallery:

        images = table.findAll("img")
        for image in images:
            try:
                pokemonName = image.get("alt")
            except:
                pass

            characters.append(pokemonName)

    return characters


characterList = generateCharacterList("https://bulbapedia.bulbagarden.net/wiki/List_of_Pokémon_by_National_Pokédex_number")

```

```python
totalChars = len(set(characterList))
pokemonDF = pd.DataFrame(columns=['Pokemon Name', 'Type','weight','Height'])
pokemonDF['Pokemon Name'] = characterList
pokemonDF.drop_duplicates(inplace=True)
pokemonDF.reset_index(inplace=True, drop=True)
pokemonDF.index += 1
pokemonDF

```

```python
characterListReduced = list(pokemonDF['Pokemon Name'][:50])


def extraInformation(pokemonList):

    for i, pokemonName in enumerate(pokemonList):

        url =  requests.get(f"https://bulbapedia.bulbagarden.net/wiki/{pokemonName}")
        soup = BeautifulSoup(url.text)

        pokemonInfoBox = soup.find_all('table', style="margin:auto; background:none;")

        types = re.findall("<b>(.*?)<", str(pokemonInfoBox[0]))
        try:
            if "Unknown" in types:
                types.pop(-1)
        except:
            types = "Unknown"
        
        pokemonInfoBox = soup.find_all('table', style="background:#FFF;")
        try:
            weight = re.findall('<td width="50%">(.*?) kg', str(pokemonInfoBox[4]))[0]
            height = re.findall('<td width="50%">(.*?) m', str(pokemonInfoBox[3]))[0]
        except:
            weight = re.findall('<td width="50%">(.*?) kg', str(pokemonInfoBox[5]))
            height = re.findall('<td width="50%">(.*?) m', str(pokemonInfoBox[4]))
            weight = [i for i in weight if float(i) != 0]
            height = [i for i in height if float(i) != 0]

        pokemonDF['Type'][i+1] = types
        pokemonDF['weight'][i+1] = weight
        pokemonDF['Height'][i+1] = height

        print(pokemonName, types, weight, height)


extraInformation(characterListReduced)    
```
