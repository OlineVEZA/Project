```python
import requests, re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import lxml
```

```python
import warnings
warnings.filterwarnings('ignore')


def generateCharacterList(url):
    websiteURL =  requests.get(url)
    soup = BeautifulSoup(websiteURL.text)
    characterGallery = soup.findAll('table',{'align':'center'})
    characters = []

    for table in characterGallery:

        images = table.findAll("img")
        for image in images:
            try:
                pokemonName = image.get("alt")
            except:
                pass

            characters.append(pokemonName)

    return characters


characterList = generateCharacterList("https://bulbapedia.bulbagarden.net/wiki/List_of_Pokémon_by_National_Pokédex_number")

```

```python
totalChars = len(set(characterList))
pokemonDF = pd.DataFrame(columns=['Pokemon Name', 'Type','Weight','Height', 'HP', 'Attack', 'Defense', 'Sp. Attack', 'Sp. Defense', 'Speed'])
pokemonDF['Pokemon Name'] = characterList
pokemonDF.drop_duplicates(inplace=True)
pokemonDF.reset_index(inplace=True, drop=True)
pokemonDF.index += 1
pokemonDF

```

```python
characterList = list(pokemonDF['Pokemon Name'])


def extraInformation(pokemonList):

    for i, pokemonName in enumerate(pokemonList):
        try:
            url =  requests.get(f"https://bulbapedia.bulbagarden.net/wiki/{pokemonName}")
            soup = BeautifulSoup(url.text)

            pokemonInfoBox = soup.find_all('table', style="margin:auto; background:none;")

            types = re.findall("<b>(.*?)<", str(pokemonInfoBox[0]))
            try:
                if "Unknown" in types:
                    types.pop(-1)
            except:
                types = "Unknown"
            
            pokemonInfoBox = soup.find_all('table', style="background:#FFF;")
            try:
                weight = re.findall('<td width="50%">(.*?) kg', str(pokemonInfoBox[4]))[0]
                height = re.findall('<td width="50%">(.*?) m', str(pokemonInfoBox[3]))[0]
            except:
                weight = re.findall('<td width="50%">(.*?) kg', str(pokemonInfoBox[5]))
                height = re.findall('<td width="50%">(.*?) m', str(pokemonInfoBox[4]))
                weight = [i for i in weight if float(i) != 0]
                height = [i for i in height if float(i) != 0]

            pokemonInfoBox = soup.find_all('div', {'class':"mw-parser-output"})
            try:
                hp = re.findall('HP<\/span><\/a>:<\/div><div style="float:right">(.*?)<\/div', str(pokemonInfoBox))
                attack = re.findall('Attack<\/span><\/a>:<\/div><div style="float:right">(.*?)<\/div', str(pokemonInfoBox))
                defense = re.findall('Defense<\/span><\/a>:<\/div><div style="float:right">(.*?)<\/div', str(pokemonInfoBox))
                spAttack = re.findall('Sp. Atk<\/span><\/a>:<\/div><div style="float:right">(.*?)<\/div', str(pokemonInfoBox))
                spDefense = re.findall('Sp. Def<\/span><\/a>:<\/div><div style="float:right">(.*?)<\/div', str(pokemonInfoBox))
                speed = re.findall('Speed<\/span><\/a>:<\/div><div style="float:right">(.*?)<\/div', str(pokemonInfoBox))

                hp, attack, defense, spAttack, spDefense, speed = str(hp[0]),str(attack[0]),str(defense[0]),str(spAttack[0]),str(spDefense[0]),str(speed[0])
            except:
                hp, attack, defense, spAttack, spDefense, speed = np.nan, np.nan, np.nan, np.nan, np.nan, np.nan
            try: 
                pokemonDF['Type'][i+1] = types
                pokemonDF['Weight'][i+1] = weight
                pokemonDF['Height'][i+1] = height
                pokemonDF['HP'][i+1] = hp
                pokemonDF['Attack'][i+1] = attack
                pokemonDF['Defense'][i+1] = defense
                pokemonDF['Sp. Attack'][i+1] = spAttack
                pokemonDF['Sp. Defense'][i+1] = spDefense
                pokemonDF['Speed'][i+1] = speed

                print(pokemonName, types, weight, height, hp, attack)
            except Exception as e:
                print(e)
                pokemonDF['Type'][i+1] = np.nan
                pokemonDF['Weight'][i+1] = np.nan
                pokemonDF['Height'][i+1] = np.nan
                pokemonDF['HP'][i+1] = np.nan
                pokemonDF['Attack'][i+1] = np.nan
                pokemonDF['Defense'][i+1] = np.nan
                pokemonDF['Sp. Attack'][i+1] = np.nan
                pokemonDF['Sp. Defense'][i+1] = np.nan
                pokemonDF['Speed'][i+1] = np.nan

                
        except:
            pass


extraInformation(characterList)   
```
